```python
import os
import PyPDF2
import re
import pandas as pd
```

```python
folder_path = 'C:/Users/User/OneDrive\Documents/SCHOOL PDFs/SEM 3.2/Mobile Programming'
df = pd.DataFrame(columns=['File', 'Topic','.1'])

keywords_topics = ['Topic']#, 'Keyword2', 'Keyword3']
keywords_1 = ['.1']#, 'Keyword2', 'Keyword3']

```

```python
# Loop through all PDF files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.pdf'):
        if file_name.startswith('MOBILE'):

            df = df.append({'File': file_name, 'Topic': 'None', '.1': "none"}, ignore_index=True)

            continue  # Skip files with prefix 'MOBILE'
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            for page_obj in pdf_reader.pages:
                text = page_obj.extract_text()
                for keyword in keywords_topics:
                    index = text.find(keyword)
                    if index != -1:
                        start_index = index + len(keyword)
                        end_index = text.find('\n', start_index)
                        if end_index == -1:
                            end_index = len(text)
                        extracted_text = text[start_index:end_index].strip()
                        df = df.append({'File': file_name, 'Topic': extracted_text}, ignore_index=True)
                
                for keyword in keywords_1:
                    index = text.find(keyword)
                    if index != -1:
                        start_index = index + len(keyword)
                        end_index = text.find('\n', start_index)
                        if end_index == -1:
                            end_index = len(text)
                        extracted_text = text[start_index:end_index].strip()
                        df = df.append({'File': file_name, '.1': extracted_text}, ignore_index=True)


# Write the DataFrame to an Excel file
df#.to_excel('output.xlsx', index=False)
```
